```python
cews = pd.read_csv("cews_updated.csv")
```

```python
# Change aggregation values

cews.loc[(cews.GeoAggregation == 'urban/rural by country'), 'GeoAggregation'] = "Urban/Rural (Canada)"
cews.loc[(cews.GeoAggregation == 'Province'), 'GeoAggregation'] = "Provinces and Territories"
cews.loc[(cews.GeoAggregation == 'urban/rural by province'), 'GeoAggregation'] = "Urban/Rural By Province/ Territory"
cews.loc[(cews.IndustryAggregation == 'All industries'), 'IndustryAggregation'] = "All Industries"
cews.loc[(cews.IndustryAggregation == 'Level 1'), 'IndustryAggregation'] = "Industry Sector (2 Digit NIACS)"
cews.loc[(cews.IndustryAggregation == 'Level 2'), 'IndustryAggregation'] = "Industry Subsector (3 Digit NIACS)"
```

```python
cews.loc[(cews["GeoAggregation"] == "Urban/Rural (Canada)") & (cews["GeographicClassification"] == "RURAL"), "Region"] = "Canada - Rural Part"
cews.loc[(cews["GeoAggregation"] == "Urban/Rural (Canada)") & (cews["GeographicClassification"] == "URBAN"), "Region"] = "Canada - Urban Part"
```
