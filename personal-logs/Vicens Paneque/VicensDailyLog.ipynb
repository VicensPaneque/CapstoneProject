```python
import pandas as pd
import numpy as np

import regex as re
import altair as alt

from datetime import datetime
from datetime import timedelta
```

```python
df = pd.read_csv("VicensDailyLog.csv")
df.head()
```

```python
one = alt.Chart(df.query("Week == '1'")).mark_bar().encode(
    y=alt.Y('Activity:N', sort='-x',
           axis=alt.Axis(titleAngle=0, titleY=-10)),
    x=alt.X('Hours', type='quantitative', aggregate='sum', title='Hours worked',
            axis=alt.Axis(format=','))
    ).properties(title={'text': 'Summary of individual work', 'subtitle': 'May 3 - 9, 2021'}
                )

two = alt.Chart(df.query("Week == '1'")).mark_bar().encode(
    x=alt.X('Hours', type='quantitative', aggregate='sum', title='Hours worked',
            axis=alt.Axis(format=','))
    ).properties(title={'text': 'Total hours', 'subtitle': 'May 3 - 9, 2021'}
                )

(two & one).configure_axis(domain=False, grid=False, ticks=False
                                ).configure_view(stroke="transparent")
```
